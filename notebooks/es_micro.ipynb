# Edge sampling - microscopic results

* We investigate the sequence of edges for a temporally growing network
* We investigate the ratio of different types of edges as the function of the size of the network

```python
stats_folder="~/networkgrowth/stats/"
```

```python
%matplotlib inline
import sys
sys.path.insert(0,"../python")
from plot.plot import Plot
from plot.edge_sampling_parser import EdgeSamplingParser
from plot.edge_sampling import EdgeSampling
Plot().set()
parser =  EdgeSamplingParser()
data = parser.run(stats_folder)
```

## Ratio of different types of edges


* Types of edges:
    * (I) influence: when one of the nodes on the new edge was already part of the network
    * (R) random: when both nodes on the new edge are new incoming nodes in the net
    * (H) homophily: when both nodes were already part of the netwowk

```python
edge_sampling = EdgeSampling(data)
edge_sampling.plot()
```
